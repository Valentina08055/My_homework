```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('googleplaystore.csv')
data.head()
# data.shape
```

```python
data = data[data['Rating']<= 5 ].sort_values('Rating', ascending=False)
```

```python
*************************************************************************************************************************
Выберем приложения с наибольшим количеством скачиваний (более 500 млн), таких 72.
Построим рейтинг по категориям. Наибольшее количество приложений среди них относятся к категориям COMMUNICATION и GAME,
и только одно приложение в данном сегменте - HEALTH_AND_FITNESS  
Совершенно другую картину мы увидим, если будем оценивать общее количество игр.
```

```python
data_most_installs = data[data['Installs'] == '500,000,000+'].sort_values('Rating', ascending=False)
# ['Installs'] == '500,000,000+' 
```

```python
len(data_most_installs)
```

```python
data_most_installs_by_category = data_most_installs.groupby('Category').count().sort_values(['App'], ascending=False)['App']
data_most_installs_by_category
```

```python
plt.title('Самые скачиваемые приложения по категориям', fontsize=17)
data_most_installs_by_category.plot(kind = 'bar')
```

```python
*******************************************************************************************************************************
```

```python
Найдем ТОП-5 категорий платных и бесплатых приложений (по количеству доступных приложений). 
```

```python
data_type_categ = data.groupby(['Type','Category']).count()['App']
```

```python
data_type_categ
```

```python
data_type_free = data_type_categ.unstack('Type').sort_values(['Free'], ascending=False).head()
# print(data_type_free['Free'].head())
```

```python
data_type_paid = data_type_categ.unstack('Type').sort_values(['Paid'], ascending=False).head()
# print(data_type_paid['Paid'].head())
```

```python
plt.figure(figsize=(5, 5))
plt.title('ТОП-5 категорий бесплатных приложений', fontsize=17)
data_type_free['Free'].plot(kind = 'pie')
# print(data_type_free['Free'])
```

```python
plt.figure(figsize=(5, 5))
plt.title('ТОП-5 категорий платных приложений', fontsize=17)
data_type_paid['Paid'].plot(kind = 'pie')
# print(data_type_paid['Paid'])
```

```python
Посмотрим распределение оценок у топовых категорий приложений.
```

```python
top_category = data.Category.value_counts().sort_values(ascending = False).head(5).index.values
top_category
ax = sns.boxplot(x="Category", y="Rating", 
            data=data[data.Category.isin(top_category)])
```

```python
******************************************************************************************************************************
```

```python
Построим heat map распределения рейтингов приложений по категориям и условным возрастным группам
```

```python
platform_genre_sales = data.pivot_table(
                        index='Category', 
                        columns='Content Rating', 
                        values='Rating', 
                        aggfunc='mean').fillna(0).applymap(float)

platform_genre_sales
```

```python
plt.figure(figsize=(10, 10))
ax = sns.heatmap(platform_genre_sales,cmap='magma',linecolor='white',linewidths=0.1)
```
