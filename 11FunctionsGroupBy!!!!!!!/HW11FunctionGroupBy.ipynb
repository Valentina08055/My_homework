#### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
	- оценка 2 и меньше - низкий рейтинг
	- оценка 4 и меньше - средний рейтинг
	- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
data_film = pd.read_csv(r'D:\Python\Files\ratings.csv')
```

```python
data_film.head()
# data_film.rating.unique()
```

```python
def class_rating(df):
    if df['rating'] <= 2:
        return  "низкий"
    if df['rating'] <= 4 and  df['rating'] >2:
        return  "средний"
    if df['rating'] > 4:
        return  "высокий"
```

```python
data_film['class'] = data_film.apply(class_rating, axis = 1)
```

Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

Результат классификации запишите в отдельный столбец region.

```python
data = pd.read_csv(r'D:\Python\Files\keywords.csv')
data.head()
```

```python
def search_town(df):
    if ('москва' or 'тула' or 'ярославль') in df.keyword:
        return "Центр"
    if ('петербург' or 'псков' or 'мурманск') in df.keyword:
        return "Северо-Запад"
    if ('владивосток' or 'сахалин' or 'хабаровск') in df.keyword:
        return "Дальний Восток"
```

```python
data['new'] = data.apply(search_town, axis = 1)
```
