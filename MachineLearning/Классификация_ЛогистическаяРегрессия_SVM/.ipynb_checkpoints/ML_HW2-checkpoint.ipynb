Делим на 2 категории: американцы и сотрудники, приехавшие из других стран

```python
def american_foreigner(data):
    if data['native-country'] == 'United-States':
        return "american"
    else:
        return "foreigner"
```

```python
data['native-country'] = data.apply(american_foreigner, axis = 1)
```

```python
data_american_foreigner = data.groupby('native-country').count().race

plt.figure(figsize=(5, 5))
plt.title('Распределение американцев и иностранных граждан', fontsize=17)
data_american_foreigner.plot(kind = 'pie')
```

```python
data_american_foreigner = data.groupby('marital-status').count().race

plt.figure(figsize=(5, 5))
plt.title('Распределение людей, живущих совместно и раздельно', fontsize=17)
data_american_foreigner.plot(kind = 'pie')
```

Преобразовываем столбец gender LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit( data['gender'] )
```

```python
le.classes_
```

```python
data['gender'] = pd.Series( data = le.transform( data['gender'] ) )
```

```python
data.head(15)
```

```python
data.gender.value_counts()
```

Преобразовываем категориальные значения OneHotEncording

```python
data_OHE = data[['workclass', 'marital-status', 'occupation', 'race', 'native-country']]
data_OHE = pd.get_dummies(data_OHE)
```

```python
data_OHE.shape
```

```python
X = data[['age', 'educational-num', 'hours-per-week', 'gender']].join(data_OHE)
```

```python
y = data['income']
```

Делим на тестовые и тренировочные

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# обучаем модель

model.fit( X_train, y_train )
predictions = model.predict_proba( X_test )
```
