По мнению рекрутеров, есть общие отличия в отношении к работе у людей, которые живут в семье и одиноких.
Одинокие больше посвящат себя работе, но чаще ее меняют. Семейные люди меньше энергии уделяют работе, но более стабильны.

Делим на 2 категории: американцы и сотрудники, приехавшие из других стран

```python
def american_foreigner(data):
    if data['native-country'] == 'United-States':
        return "american"
    else:
        return "foreigner"
```

```python
data['native-country'] = data.apply(american_foreigner, axis = 1)
```

```python
data_american_foreigner = data.groupby('native-country').count().race

plt.figure(figsize=(5, 5))
plt.title('Распределение американцев и иностранных граждан', fontsize=17)
data_american_foreigner.plot(kind = 'pie')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
label_gender = le.fit_transform(data['gender'])
data.loc[:, 'label_gender'] = label_gender

label_relationship = le.fit_transform(data['relationship'])
data.loc[:, 'label_relationship'] = label_relationship

label_marital_status = le.fit_transform(data['marital-status'])
data.loc[:, 'label_marital_status'] = label_marital_status

label_income = le.fit_transform(data['income'])
data.loc[:, 'label_income'] = label_income
data
```
