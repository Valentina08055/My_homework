```python
import re

# palette = '|'.join(color_list) 
# data['animal_colors']= data['color'].str.extract('(palette)')
data['animal_colors']= data['color'].str.extract('(Black|Brown|White|Tan|Orange|Blue|Tricolor|Tortie|Calico|Red|Torbie)')
data['animal_colors'] = data['animal_colors'].fillna("other_color")
data.head()
```

Преобразуем даты в формат дат, посчитаем возраст животного в годах.

```python
from datetime import datetime, date, time, timedelta
from pandas import DataFrame
```

```python
data['date_of_birth'] = pd.to_datetime(data['date_of_birth'])
data['monthyear'] = pd.to_datetime(data['monthyear'])
data['animal_age'] = (data['monthyear'] - data['date_of_birth'])
```

```python
data['animal_age'] = round(data['animal_age'] / np.timedelta64(1,'Y'))
```

```python
# data_breed = data.groupby('breed').count()
# data_breed.sort_values(['animal_type'],ascending=False).head(20)

data_breed = data.groupby(by=['breed'], as_index=False)['animal_type'].count()
data_breed = data_breed.sort_values(['animal_type'],ascending=False)
data_breed.head(20)
```

```python
list_breed = data_breed['breed'][:20].tolist()
list_breed 
```

```python
def add_breeds(data):
    for i in list_breed:
        if str(i) in data.breed:
            return i
        else: 'other'
```

```python
data['animal_breed']= data.apply(add_breeds, axis = 1)
```

```python
data.head()
```

```python
data['animal_breed'] = data['animal_breed'].fillna("other")
```

```python
ohe_monthyear = oh_enc.fit_transform(data[['monthyear']]).toarray()
ohe_monthyear = pd.DataFrame(ohe_monthyear, columns=oh_enc.categories_[0])
ohe_monthyear.head()
```

```python
ohe_sex_upon_outcome = oh_enc.fit_transform(data[['sex_upon_outcome']]).toarray()
ohe_sex_upon_outcome = pd.DataFrame(ohe_sex_upon_outcome, columns=oh_enc.categories_[0])
ohe_sex_upon_outcome.head()
```

```python
ohe_animal_colors = oh_enc.fit_transform(data[['animal_colors']]).toarray()
ohe_animal_colors = pd.DataFrame(ohe_animal_colors, columns=oh_enc.categories_[0])
ohe_animal_colors.head()
```

```python
ohe_animal_breed = oh_enc.fit_transform(data[['animal_breed']]).toarray()
ohe_animal_breed = pd.DataFrame(ohe_animal_breed, columns=oh_enc.categories_[0])
ohe_animal_breed.head()
```
